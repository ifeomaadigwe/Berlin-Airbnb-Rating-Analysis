# Data Cleaning!!!

```python
def clean_data(df):
    for column in df.columns:
        if df[column].dtype == 'object':
            df[column].fillna(df[column].mode()[0], inplace=True)
        else:
            df[column].fillna(df[column].mean(), inplace=True)
    
    df.drop_duplicates(inplace=True)
    return df

cleaned_data = clean_data(df)
#print(cleaned_data)
```

```python
# To check the unique values in a specific column

unique_values_in_column = df['Price'].unique()
#print(unique_values_in_column)
```

```python
# To check the actually values in the column

column_name = df['Is Exact Location'] 
column_name = column_name.tolist()
#print(column_name)
```

```python
#The following columns contain undiscribable values('f' and 't')replace f with false and t with true

columns_to_clean = ['Instant Bookable', 'Business Travel Ready', 'Is Superhost', 'Is Exact Location']

for column in columns_to_clean:
    df[column] = df[column].replace('f', False)
    df[column] = df[column].replace('t', True)
```

```python
# Remove irrelevant columns

df.drop(['Listing ID', 'Listing URL', 'Host URL', 'Host ID'], axis = 1, inplace = True)
```

```python
# Remove irrelevant rows

row_to_delete = 1 
df.drop(row_to_delete, inplace=True)
```

```python
# The'Price' column contains non-numeric values, such as currency $ symbols or commas.
# To clean this column:

df['Price'] = df['Price'].str.replace('$', '').str.replace(',', '').astype(float)
```

```python
# Check and change the data type of some columns to enable calculations and aggregations

df['Review_Date'] = pd.to_datetime(df['review_date'], errors='coerce')
df['Host_since'] = pd.to_datetime(df['Host Since'], errors='coerce')
df['Host_response_time'] = pd.to_timedelta(df['Host Response Time'], errors='coerce')
df['Host_response_rate'] = pd.to_numeric(df['Host Response Rate'], errors='coerce')
df['Price$'] = pd.to_numeric(df['Price'], errors='coerce')
df['First_review'] = pd.to_datetime(df['First Review'], errors='coerce')
df['Last_review'] = pd.to_datetime(df['Last Review'], errors='coerce')
```

### Price Fluctuations caused by different seasons

```python
df.head(1)
```

```python
# Extract and create a Month Column

df['Month'] = df['Review_Date'].dt.month  ## from the 'review_date' column
#df.head(1)
```

```python
# Group the months to season names

df['Season'] = np.where(df['Month'] <= 2, 'Winter', 
                        np.where(df['Month'] <= 5, 'Spring',
                                 np.where(df['Month'] <= 8, 'Summer',
                                          np.where(df['Month'] <= 11, 'Fall', 'Winter'))))

#print(df['Season'])
```

```python
####  ANOTHER METHOD TO GROUP AND CREATE A SEASON COLUMN

# Group the months to season names

season_mapping = {
    12: 'Winter', 1: 'Winter', 2: 'Winter',
    3: 'Spring', 4: 'Spring', 5: 'Spring',
    6: 'Summer', 7: 'Summer', 8: 'Summer',
    9: 'Fall', 10: 'Fall', 11: 'Fall'
}

# create a new 'Season' column

df['Season'] = df['Month'].map(season_mapping)
```

```python
# Average price by season

average_price_by_season = df.groupby('Season')['Price'].mean().reset_index()
average_price_by_season
```

```python
# Average price by season - pie chart

average_price_by_season = df.groupby('Season')['Price'].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.pie(average_price_by_season['Price'], labels=average_price_by_season['Season'], autopct='%1.1f%%', startangle=140)
plt.title('Average price by season')
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular.
plt.show()
```

```python
# Average price by season - Bar chart

average_price_by_season = df.groupby('Season')['Price'].mean().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(x='Season', y='Price', data=average_price_by_season, palette='viridis')
plt.xlabel('Season')
plt.ylabel('Average Price')
plt.title('Seasonal Price Fluctuations in Berlin Airbnb Listings')
plt.show()
```
